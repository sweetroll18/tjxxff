```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# 创建数据集
dataset = np.random.rand(100,2) # 随机出一个数据集
#print(dataset)
def trueFunction(a): # 定义一个真正的函数y=0.9x+0.1
    return 0.9*a+0.1
label = np.sign(dataset[:,1]-trueFunction(dataset[:,0])).reshape(-1,1) # 在直线上方的是1，下方的标注为-1。并将数据整成一列。
#print(label)
dataset[1]
```

```python
# 定义感知机
class Perception:
    def __init__(self, lr=0.005, train_maxnum=100):
        # w_size：元组
        self.w = np.zeros(2) 
        self.b = 0.0
        self.lr = lr
        self.maxnum = train_maxnum
    def function(self, x):
        return self.w*x + self.b
    def train(self, X, Y):
        times = 0
        print('-1-')
        while times<self.maxnum:
            #print('-2-')
            wrong_num = 0
            for i in range(len(X)):
                #print('-3-')
                if Y[i]*(np.dot(self.w, X[i])+self.b) <= 0:
                    #print('-4-')
                    self.w = self.w + self.lr * Y[i] * X[i]
                    self.b = self.b + self.lr * Y[i]
                    wrong_num = wrong_num + 1
            if wrong_num == 0:
                print('-5-')
                break
            times = times + 1
            #print(self.w, self.b)
            #print('-6-')
perception = Perception()
perception.train(dataset, label)
```

```python
plt.scatter(dataset[:,0], dataset[:,1], label='0')
x_points = np.linspace(0.0, 1.0, 10)
y_ = 0.9 * x_points + 0.1
plt.plot(x_points, y_, label='True')
y__ = -(perception.w[0]/perception.w[1]) * x_points -(perception.b/perception.w[1]) # P41的笔记推导
plt.plot(x_points, y__, label='Train')
plt.legend()
```
